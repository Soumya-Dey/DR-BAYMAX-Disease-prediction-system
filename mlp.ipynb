```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data/formatted_health_data.csv")
ds = pd.read_csv('data/Health_dataset.csv', header=None, dtype=str, keep_default_na=False)
ds.drop(index = 0, axis = 0, inplace = True)
ds.reset_index(inplace = True, drop = True)
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
data.drop(columns=['Unnamed: 73'], axis = 1)
data.head()
```

```python
for col in data.columns:
    data = data.rename(columns={col:col.strip()})
```

```python
training_data = data.iloc[:, :-1]
training_labels = data['disease']
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set, train_labels, test_labels = train_test_split(training_data, training_labels, random_state =0, test_size= 0.30)
```

```python
print('Training Set: ', len(train_set))
print('Test Set: ', len(test_set))
print('Training Labels: ', len(train_labels))
print('Test Labels: ', len(test_labels))
```

```python
train_dist = train_labels.value_counts()
```

```python
test_dist = test_labels.value_counts()
```

```python
all_disease = training_labels.unique()
```

```python
disease_dict = {"Disease": [], "Training": [], "Testing": []}
for d in all_disease:
    disease_dict['Disease'].append(d)
    disease_dict['Training'].append(train_dist[d])
    disease_dict['Testing'].append(test_dist[d])
```

```python
disease_dist_df = pd.DataFrame(disease_dict)
disease_dist_df
disease_dist_df.to_csv('./data/disease_distribution.csv', index=False)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
svm = MLPClassifier()
svm.fit(train_set, train_labels)
```

```python
predictions = svm.predict(test_set)
```

```python
test_labels.iloc[517]
```

```python
predictions[517]
```

```python
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
accuracy_score(test_labels, predictions)
```

```python
report = classification_report(test_labels, predictions, output_dict=True)
```

```python
print(classification_report(test_labels, predictions))
```
